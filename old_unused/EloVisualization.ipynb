```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
import ast
```

```python
df_elo_skipper = pd.read_csv("elo15.csv")
df_elo_crew = pd.read_csv("elo16.csv")
df_races = pd.read_csv("races.csv")

f24_sailors = df_races.loc[df_races['raceID'].str.contains('f24'),'Sailor'].unique()

df_elo_skipper_24 = df_elo_skipper.loc[df_elo_skipper["Sailor"].isin(f24_sailors)].reset_index(drop=True)
df_elo_crew_24 = df_elo_crew.loc[df_elo_crew["Sailor"].isin(f24_sailors)].reset_index(drop=True)
df_elo_crew_24
```

```python
df_sailor = df_elo_crew_24[df_elo_crew_24['Sailor'].str.contains("Elliott Chalcraft")]
df_sailor = df_sailor.dropna(axis=1, how='all')

df_sailor2 = df_elo_skipper_24[df_elo_skipper_24['Sailor'].str.contains("Carter Anderson")]
df_sailor2 = df_sailor2.dropna(axis=1, how='all')

df_sailor
```

```python
df_temp = df_sailor.drop(['Sailor', 'Team', 'Pos', 'Elo'], axis='columns')
df_temp2 = df_sailor2.drop(['Sailor', 'Team', 'Pos', 'Elo'], axis='columns')
fig = go.Figure()
# df_temp2 = pd.DataFrame({
#     'Column': df_temp.columns,
# })

fig.add_trace(go.Bar(x=df_temp.columns, y=[float(ast.literal_eval(s)[1]) for s in df_temp.iloc[0]],
                name="Elliott"))
fig.add_trace(go.Bar(x=df_temp2.columns, y=[float(ast.literal_eval(s)[1]) for s in df_temp2.iloc[0]],
                name='Carter'))
# fig.add_trace(go.Bar(x=years, y=[300, 400, 700],
#                 base=0,
#                 marker_color='lightslategrey',
#                 name='revenue'
#                 ))


    # df_temp2[f'Values{i}']
# fig.add_trace(go.Scatter(x=df_temp.columns, y=[float(ast.literal_eval(s)[i]) for s in df_temp.iloc[0]], mode='lines', name=['score','change','opponents','elo'][i]))

        
fig.update_layout(
    title='Scores and changes',
    xaxis_title='Regatta',
    yaxis_title='Elo',
    height=300, width=900
)
fig.update_xaxes(showticklabels=False)
fig.show()


fig = go.Figure()

for i in range(2,4):
    # df_temp2[f'Values{i}']
    fig.add_trace(go.Scatter(x=df_temp2.columns, y=[float(ast.literal_eval(s)[i]) for s in df_temp2.iloc[0]], mode='lines', name=['score','change','opponents','elo'][i]))

    # plt.line(df_temp2, x='Column', y='Value')
    
fig.update_layout(
    title='Elo change over time',
    xaxis_title='Regatta',
    yaxis_title='Elo',
    height=300, width=2500
)
# fig.update_xaxes(showticklabels=False)
fig.show()
```
